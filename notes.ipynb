# MLOps Vehicle Insurance Project - `src/` Folder Structure

## Overview
The `src/` directory contains all core project code organized in a modular, scalable structure. Each folder has a specific responsibility in the ML pipeline.

---

## Folder Breakdown

### **components/**
- **Purpose:** Contains the main ML pipeline components
- **Real-world use:** Each component represents a reusable step in the workflow
- **Key files:**
  - `data_ingestion.py` - Fetches raw data from sources
  - `data_validation.py` - Validates data quality and schema
  - `data_transformation.py` - Preprocesses and engineers features
  - `model_trainer.py` - Trains the ML model
  - `model_evaluation.py` - Evaluates model performance
  - `model_pusher.py` - Deploys model to production

### **entity/**
- **Purpose:** Defines data structures and classes (objects passed between components)
- **Real-world use:** Creates strongly-typed configurations and artifacts for type safety
- **Key files:**
  - `config_entity.py` - Configuration objects (input parameters for each component)
  - `artifact_entity.py` - Artifact objects (output results from each component)
  - `estimator.py` - Model handling and predictions
  - `s3_estimator.py` - Cloud model storage/retrieval

### **configuration/**
- **Purpose:** Manages connections to external services
- **Real-world use:** Centralizes credentials and connection logic
- **Key files:**
  - `mongo_db_connection.py` - MongoDB connection setup and queries
  - `aws_connection.py` - AWS S3 and other cloud services

### **constants/**
- **Purpose:** Stores all hardcoded constant values
- **Real-world use:** Single source of truth for config values (prevents magic numbers)
- **Contents:** Database names, API keys, model paths, threshold scores, bucket names

### **data_access/**
- **Purpose:** Data layer - abstracts database queries
- **Real-world use:** Isolates database logic from business logic
- **Key files:**
  - `proj1_data.py` - Fetches data from MongoDB and converts to DataFrame

### **cloud_storage/**
- **Purpose:** Manages cloud service interactions
- **Real-world use:** Handles model and artifact uploads/downloads to production
- **Key files:**
  - `aws_storage.py` - S3 bucket operations

### **logger/**
- **Purpose:** Custom logging system
- **Real-world use:** Tracks execution flow, debugging, and monitoring in dev/prod environments

### **exception/**
- **Purpose:** Custom exception classes
- **Real-world use:** Enables granular error handling specific to your application

### **pipline/** (Note: spelled as "pipline" in the project)
- **Purpose:** Orchestrates the entire workflow
- **Real-world use:** Controls execution order of components
- **Key files:**
  - `training_pipeline.py` - Orchestrates model training workflow
  - `prediction_pipeline.py` - Handles inference/predictions in production

### **utils/**
- **Purpose:** Helper functions and utilities
- **Real-world use:** Shared functions used across multiple components
- **Key files:**
  - `main_utils.py` - Common operations (data processing, model saving/loading)

---

## Architecture in Action

```
Raw Data (MongoDB)
    ↓
[data_access] → Fetch data
    ↓
[components/data_ingestion] → Load data
    ↓
[components/data_validation] → Validate data
    ↓
[components/data_transformation] → Preprocess & engineer features
    ↓
[components/model_trainer] → Train model
    ↓
[components/model_evaluation] → Evaluate performance
    ↓
[components/model_pusher] → Push to AWS S3
    ↓
Production Ready Model
```

All orchestrated by `pipline/training_pipeline.py` using entity configs and artifacts.

## Why we include `-e .` in `requirements.txt`

`-e .` means **install the current project as an editable package**.

### What it does
- Installs your local project (the folder with `setup.py` or `pyproject.toml`) into the environment.
- Uses **editable mode**, so code changes in your project are immediately reflected without reinstalling.

### Why it is useful in this project
- Lets all modules under `src/` be imported cleanly across notebooks, scripts, and pipeline files.
- Supports active development (frequent code updates).
- Keeps dependency setup consistent for training, evaluation, and deployment workflows.

### In short
Without `-e .`, Python may not recognize your local package properly, and you may need repeated reinstalls after code changes.
